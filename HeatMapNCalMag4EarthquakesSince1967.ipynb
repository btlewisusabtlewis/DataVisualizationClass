```
import pandas as pd
```

```
ncal_mag4 = pd.read_csv('https://raw.githubusercontent.com/btlewisusabtlewis/sample_data/master/Mag4PlusNCalEarthquakesSince1967.csv')
```

```
ncal_mag4.head()
```

```
ncal_mag4.info()
```

```
just_lat_long = ncal_mag4[['Latitude', 'Longitude']]
```

```
just_lat_long.head()
```

```
ncal_mag4_lat_long = just_lat_long.to_numpy()
```

```
ncal_mag4_lat_long.size
```

```
ncal_mag4_lat_long.shape
```

```
import folium
from folium import plugins
```

```
m = folium.Map([38.0	, -120.8958], zoom_start=8)
```

```
m.add_child(plugins.HeatMap(ncal_mag4_lat_long))
```

```
BIG_MAGNITUDE = 5.5
ncal_big = ncal_mag4[ncal_mag4.Magnitude > BIG_MAGNITUDE]
```

```
ncal_big.head()
```

```
ncal_big.info()
```

```
from datetime import datetime
```

```
for index, row in ncal_big.iterrows():
  (dt, mSecs) = row['DateTime'].strip().split(".") 
  date_and_time = datetime.strptime(dt, '%Y/%m/%d %H:%M:%S')
  date_and_time_str = date_and_time.strftime('%m/%d/%Y')
  mag = row['Magnitude']
  if mag > 6.0:
    icon_color = 'red'
  elif mag > 5.8:
    icon_color = 'orange'
  else:
    icon_color = 'lightgray'
  folium.Marker([row['Latitude'], row['Longitude']], 
                popup=date_and_time_str, 
                tooltip=('Mag ' + str(mag)), 
                icon=folium.Icon(color=icon_color, icon='exclamation-triangle', prefix='fa')).add_to(m)
```

```
m
```

```
m2 = folium.Map([38.0	, -120.8958], zoom_start=8)
m2.add_child(plugins.HeatMap(ncal_mag4_lat_long))
```

```
for index, row in ncal_big.iterrows():
  (dt, mSecs) = row['DateTime'].strip().split(".") 
  date_and_time = datetime.strptime(dt, '%Y/%m/%d %H:%M:%S')
  date_and_time_str = date_and_time.strftime('%m/%d/%Y')
  mag = row['Magnitude']
  if mag > 6.0:
    rad = 8000
  elif mag > 5.8:
    rad = 5000
  else:
    rad = 3700
  folium.Circle(
    radius=rad,
    location=[row['Latitude'], row['Longitude']],
    tooltip=('Mag ' + str(mag) + ' ' + date_and_time_str),
    color='black',
    fill=True,
    fill_color='gray',
    fill_opacity=0.15
  ).add_to(m2)

m2
```

```
m.save('NCalEarthquakesSince1967_Markers.html')
```

```
m2.save('NCalEarthquakesSince1967_Circles.html')
```
